```python
# import library to run script

import html2text
from bs4 import BeautifulSoup
# from urllib.request import urlopen
import itertools
import requests
```

```python
# prepare targets and declare variables

url = "https://tirupati.emunicipal.ap.gov.in/ptis/view/viewProperty-viewForm.action?propertyId=**********"

html = requests.get(url)
# print(html.status_code)

# html = urlopen(url).read()
soup = BeautifulSoup(html.content)

# Remove unwanted tags for my case i don't need below tags, no useful information onn  it

for script in soup(["script", "style", "head", "header", "footer", "button"]):
    script.decompose()

strips = list(soup.stripped_strings)
```

```python

keywordsOf = dict() 
for index,value in enumerate(strips):
  keywordsOf[index] = value
print(keywordsOf)
```

```python
# prints list of strings after removal of all html tags
# This strip is used for keyword based search
strips
```

```python
# gets information in div's where class name is "col-xs-3 add-margin"
l = soup.find_all(attrs={"class": "col-xs-3 add-margin"})

# gets information in div's where class name is "col-xs-3 add-margin view-content"

m = soup.find_all(attrs={"class": "col-xs-3 add-margin view-content"})
```

```python
# convert list to html and preparing list
l = BeautifulSoup(str(l), 'html.parser')
m = BeautifulSoup(str(m), 'html.parser')
l = list(l.stripped_strings)
m = list(m.stripped_strings)
```

```python
# storing property details in dictionary

PropertyDetails = {}

for (a, b) in itertools.zip_longest(l,m):
    PropertyDetails[a] = b
```

```python
# print property details

PropertyDetails
```

```python
# gets floor details

d = soup.find_all(attrs={"id": "floorDetails"})

d = BeautifulSoup(str(d), 'html.parser')

e = d.find_all(attrs={"class": "blueborderfortd"})

f = d.find_all(attrs={"class": "bluebgheadtd"})


e = BeautifulSoup(str(e), 'html.parser')
f = BeautifulSoup(str(f), 'html.parser')
e = list(e.stripped_strings)
f = list(f.stripped_strings)

```

```python
# get floor counts

g = d.find_all(attrs={"id": "Floorinfo"})

# store floor details

my_floor_details = {}
h = 0
while h < len(g):
  for (a, b) in itertools.zip_longest(f,e):
    a = str(a) + "-" +str(h)
    my_floor_details[a] = b

  h=h+1
```

```python
# display floor details
my_floor_details
```

```python
#gets owner information in divs 

OwnerNameDiv = soup.find_all(attrs={"id": "OwnerNameDiv"})
OwnerNameDiv = BeautifulSoup(str(OwnerNameDiv), 'html.parser')
nameRow_id_count = len(OwnerNameDiv.find_all(attrs={"id": "nameRow"}))


OwnerData = OwnerNameDiv.find_all(attrs={"class": "blueborderfortd"})

OwnerDataHeadings = OwnerNameDiv.find_all(attrs={"class": "bluebgheadtd"})


OwnerDataHeadings = BeautifulSoup(str(OwnerDataHeadings), 'html.parser')
OwnerData = BeautifulSoup(str(OwnerData), 'html.parser')
OwnerDataHeadings = list(OwnerDataHeadings.stripped_strings)
OwnerData = list(OwnerData.stripped_strings)

owner_details = {}
h = 0
while h < nameRow_id_count:
  for (a, b) in itertools.zip_longest(OwnerDataHeadings,OwnerData):
    a = str(a) + "-" +str(h)
    owner_details[a] = b

  h=h+1
```

```python
# display owner details

owner_details
```
